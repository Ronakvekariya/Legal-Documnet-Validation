Testing the annoatation for the adhar card

```python
# Define the path to the text file
text_file_path = "./aadhar card.v2i.yolov8/valid\labels/427324185891_back_jpg.rf.52dadda62b57c241216716bb677a61f3.txt"

# Read the content of the text file
with open(text_file_path, 'r') as file:
    text_content = file.read()

# Print the content of the text file
coor = text_content.split()

# Convert annotation values to floats
_, center_x, center_y, width, height = map(float, coor)
```

```python
# Load the image
image_path = "./aadhar card.v2i.yolov8/valid\images/427324185891_back_jpg.rf.52dadda62b57c241216716bb677a61f3.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib


# Image dimensions
img_height, img_width, _ = image.shape

# Denormalize the coordinates
center_x = int(center_x * img_width)
center_y = int(center_y * img_height)
width = int(width * img_width)
height = int(height * img_height)

# Calculate top-left and bottom-right corners of the bounding box
x1 = int(center_x - width / 2)
y1 = int(center_y - height / 2)
x2 = int(center_x + width / 2)
y2 = int(center_y + height / 2)

# Draw the bounding box on the image
cv2.rectangle(image, (x1, y1), (x2, y2), color=(255, 0, 0), thickness=2)

# Display the image with the bounding box
plt.imshow(image)
plt.axis("off")
plt.show()
```

Now changing the label of pan card from 0 to 1

```python
for folder in os.listdir('./PAN Model.v1i.yolov8'):
    if os.path.isdir('./PAN Model.v1i.yolov8/' + folder):
        for images in os.listdir('./PAN Model.v1i.yolov8/' + folder + '/labels'):
                print(images)
                normalized_lines = []
                if os.path.exists('./PAN Model.v1i.yolov8/' + folder + '/labels/' + images):
                    print("exists")
                    with open('./PAN Model.v1i.yolov8/' + folder + '/labels/' + images, 'r') as label_file:
                        lines = label_file.readlines()
                        print(lines)

                    # Loop through each bounding box and plot it
                    for line in lines:
                        parts = line.strip().split()
                        class_id = int(parts[0])
                        class_id = "1"
                        x_center, y_center, width, height = map(float, parts[1:])

                        # Create the normalized label line in YOLO format
                        normalized_line = f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n"
                        normalized_lines.append(normalized_line)

                    if normalized_lines:
                        print(normalized_lines)
                        with open('./PAN Model.v1i.yolov8/' + folder + '/labels/' + images, 'w') as label_file:
                            label_file.writelines(normalized_lines)
                    else:
                        print("the file is alright")
    
```
